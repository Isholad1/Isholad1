## Distribution


### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


Plot the distribution curve for the column `ConvertedComp`.


```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Assuming the "ConvertedComp" column contains salaries in USD
# Plot the distribution curve
plt.figure(figsize=(10, 6))
sns.distplot(df['ConvertedComp'], bins=50, kde=True, color='skyblue')

# Set plot labels and title
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Converted Compensation')

# Show the plot
plt.show()

```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
# Assuming the column containing converted compensation is named 'ConvertedComp'
# Plot a box plot to identify outliers
plt.figure(figsize=(10, 6))
plt.boxplot(df['ConvertedComp'], vert=False, sym='b.')

# Set plot labels and title
plt.xlabel('Converted Compensation (USD)')
plt.title('Box Plot of Converted Compensation to Identify Outliers')

# Show the plot
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# Calculate the median ConvertedComp before removing outliers
median_before_outliers = df['ConvertedComp'].median()

# Print the result
print(f"The median ConvertedComp before removing outliers is: {median_before_outliers}")
```

```python
# Calculate the interquartile range (IQR)
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Calculate upper and lower bounds
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Create a new DataFrame without outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Calculate the median ConvertedComp after removing outliers
median_after_outliers = df_no_outliers['ConvertedComp'].median()

# Print the result
print(f"The median ConvertedComp after removing outliers is: {median_after_outliers}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")

# Assuming the column containing age information is named 'Age'
# Create a boxplot for 'Age'
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Age'])

# Calculate Q1 and IQR
Q1 = df['Age'].quantile(0.25)
IQR = df['Age'].quantile(0.75) - Q1

# Calculate the lower bound for outliers below Q1
lower_bound_Q1 = Q1 - 1.5 * IQR

# Identify and count outliers below Q1
outliers_below_Q1 = df[df['Age'] < lower_bound_Q1]
num_outliers_below_Q1 = len(outliers_below_Q1)

# Print the result
print(f"The number of outliers below Q1 is: {num_outliers_below_Q1}")

# Show the boxplot
plt.show()
```

```python
# Calculate the interquartile range (IQR)
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Calculate upper and lower bounds
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Create a new DataFrame without outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Calculate the mean ConvertedComp after removing outliers
mean_after_outliers = df_no_outliers['ConvertedComp'].mean()

# Print the result
print(f"The mean ConvertedComp after removing outliers is: {mean_after_outliers}")
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
# Assuming the column containing converted compensation is named 'ConvertedComp'
# Calculate the interquartile range (IQR)
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Calculate upper and lower bounds
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Identify outliers
outliers = df[(df['ConvertedComp'] > upper_bound) | (df['ConvertedComp'] < lower_bound)]

# Count the number of outliers
num_outliers = len(outliers)

# Print the result
print(f"The number of outliers in the 'ConvertedComp' column is: {num_outliers}")
```
